```python
import pandas as pd
```

```python
data = pd.read_csv('demog_2008_onward.csv')
```

```python
data.head()
```

```python
data['year'] = '2008'
```

```python
data['city_name'] = pd.np.where(data.PLACE.str.contains("city"),
                                                         data.PLACE.str.split("city"),
                                             pd.np.where(data.PLACE.str.contains("CDP"),
                                                         data.PLACE.str.split("CDP"),
                                             pd.np.where(data.PLACE.str.contains("town"),
                                                         data.PLACE.str.split("town"),
                                             pd.np.where(data.PLACE.str.contains("(U)"),
                                                         data.PLACE.str.split("(U)"),
                                                         data.PLACE))))
```

```python
data['city_name'] = data['city_name'].str[0]
```

```python
data
```

```python
!curl https://raw.githubusercontent.com/fitnr/addfips/master/addfips/data/states.csv?accessType = Download > states.csv
```

```python
states = pd.read_csv('states.csv')
states.head()
```

```python
df = data.merge(states, left_on='STATE', right_on='name', how = 'inner')
```

```python
df.head()
```

```python
df['abrv_city_state'] = df['city_name'] + '_' + df['postal']
```

```python
df['abrv_city_state'] = df['abrv_city_state'].str.replace(' ','')
df['abrv_city_state'] = df['abrv_city_state'].str.replace(',','')
df['abrv_city_state'] = df['abrv_city_state'].str.replace('.','')
df['abrv_city_state'] = df['abrv_city_state'].str.lower()
```

```python
df.head()
```

```python
final_df = df.drop(['﻿NHGISCODE','GJOIN2010','STATE','PLACE','year','city_name','name','postal','fips'], axis = 1)
```

```python
final_df.head()
```

```python
final_df.to_csv('2008_demographics_finished.csv')
```
